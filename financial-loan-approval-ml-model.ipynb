# Introduction

## Defination On Bussiness Task

**Primary Task Topic**

This project aims to address a critical business challenge faced by a growing global banking company: the inefficiency of its manual loan application review process. The high volume of loan requests has overwhelmed the loan team, leading to delays that result in lost revenue for the bank. As a data scientist, I will develop and deploy a highly accurate machine learning model to automate the loan approval process, classifying applicants as either approved or rejected. The primary objective is to create a solution that not only streamlines operations but also maximizes profitability by accurately identifying low-risk borrowers.

**Business Problem**

A significant surge in loan applications has strained the current manual review process at our banking institution. This labor-intensive process, which relies on a team of human analysts, is proving to be a bottleneck that prevents the bank from making timely decisions. As a result, the company is unable to capitalize on potential interest revenue from qualified applicants and is exposed to increased risk from potential defaults. The core business problem is the need for a scalable, efficient, and reliable system to handle the increasing demand for loan services while mitigating financial risk.

**Proposed Solution**

To solve this problem, I will construct a supervised machine learning model using historical data gathered from the finance team. This model will be trained to learn the intricate patterns and criteria that have historically led to successful loan repayments. By leveraging this past data, the model will be able to accurately predict the likelihood of an applicant defaulting. The outcome will be a robust predictive tool that can determine whether a loan application should be approved or rejected, thereby providing the loan team with a powerful, data-driven decision-making for financial aid.

**Anticipated Business Impact**

The implementation of this machine learning model is expected to deliver significant benefits to the company. First, by automating the loan classification process, the bank can dramatically reduce operational costs associated with manual labor and reallocate valuable employee resources to other strategic tasks. Second, the model will enable faster loan approvals, allowing the bank to capture interest revenue in a more timely manner and gain a competitive advantage in the market. Most importantly, the model's ability to accurately reject high-risk applicants will directly mitigate financial losses from loan defaults, contributing to the bank's long-term financial stability and profitability.

## Dealing with missing values

**Imputation Strategies**

The dataset is found to have missing values across several columns. The approach to handling these missing values is determined by their distribution and the nature of the data.

- Target Variable Imputation: The loan_status column is the target variable for the predictive model. Given the critical role of this feature, any rows with missing values in this column are dropped. This approach prevents distortion of the model's predictions, as imputing the target variable could lead to biased results. This action results in a loss of approximately 1.6% of the data, a small price for maintaining model integrity.

- Missing Value Imputation for Categorical Features: The education column contains missing values. A visual inspection of its value distribution reveals a near-uniform split between 'Graduate' and 'Not Graduate' categories. Consequently, the missing values are imputed using the mode (the most frequent value). This method preserves the original distribution of the data without introducing bias.

- Missing Value Imputation for Numerical Features: The numerical features loan_amount, loan_term, and luxury_assets_value also contain missing values. To determine the most appropriate imputation strategy, the skewness of each column's distribution is calculated.

  - For loan_amount and luxury_assets_value, the data exhibits low skewness (<1), indicating a near-symmetrical distribution. Therefore, the mean is used to fill the missing values, as it is a robust measure of central tendency for such distributions.

  - For loan_term, although the skewness is low, the values are discrete rather than continuous. To preserve the proportional distribution of the existing data, missing values are filled by randomly sampling from the unique values in the column, weighted by their frequency. This maintains the integrity of the data's original distribution.

After all imputations, the final number of missing values is verified to confirm that the dataset is now complete.

## Oversampling

In this section, I will address the class imbalance observed in the target variable, as previously visualized in the histogram plot. To mitigate this issue, the Synthetic Minority Over-sampling Technique (SMOTE) will be applied. SMOTE is a robust oversampling method that generates synthetic data points for the minority class, thereby balancing the dataset and improving the model's ability to learn from and accurately predict the underrepresented class.

**Theory Behind Oversampling (SMOTE)**

Class imbalance is a common challenge in machine learning, where the number of instances in one class significantly outweighs the instances in other classes. This can cause a model to become biased toward the majority class, leading to poor predictive performance on the minority class. To address this issue, oversampling techniques are used to artificially increase the number of instances in the minority class, thereby balancing the dataset.

The Synthetic Minority Over-sampling Technique (SMOTE) is a popular oversampling algorithm. Instead of simply duplicating existing instances, SMOTE generates new, synthetic data points for the minority class. It works by:
- Selecting a random instance from the minority class.
- Finding its k-nearest neighbors (where 'k' is a user-defined parameter) in the feature space.
- Generating a new, synthetic instance at a random point along the line segments connecting the selected instance and its neighbors.

This process creates new data that is a blend of existing minority class instances, expanding the decision boundary of the minority class.

**Application**

In this section, the SMOTE algorithm will be applied to address the class imbalance in the target variable, as previously identified by the histogram plot. This process will create a more balanced training dataset, enabling the model to learn from both the majority and minority classes more effectively and improving its overall generalization performance.

## Theoretical Background

**K-Fold Cross-Validation**

K-fold cross-validation is a robust technique used to evaluate a model's performance and generalization ability. The process involves partitioning the entire dataset into K equal-sized folds. The model is then trained K times. In each iteration, a unique fold is reserved as the validation set, while the remaining K-1 folds are used for training. The performance metric is calculated for each of the K validation runs, and the final performance of the model is reported as the average of these K scores. This method provides a more reliable estimate of a model's performance on unseen data by reducing the variance associated with a single train-test split. K-Fold Cross-Validation is implemented by using cross_val_score from tensorflow library.

**Classification Models**

- Logistic Regression: A linear model used for binary classification. It estimates the probability of a data point belonging to a particular class by fitting a logistic function to the data.

- Naive Bayes: Based on Bayes' theorem, this is a probabilistic classifier that assumes a strong (naive) independence between features. It's often used for tasks like spam filtering and text classification due to its simplicity and efficiency.

- Support Vector Machine (SVM): A model that finds the optimal hyperplane in a high-dimensional space to separate data points into different classes. It aims to maximize the margin between the classes to improve classification accuracy.

- K-Nearest Neighbor (k-NN): A non-parametric, instance-based learning algorithm. It classifies a new data point by a majority vote of its k nearest neighbors, where k is a user-defined integer. The class assigned is the most common among its neighbors.

- Decision Tree: A tree-like model that makes decisions by splitting the data into subsets based on the most significant features. The goal is to create a tree with clear, distinct leaf nodes that represent the final classification.

**Ensemble Learning Models**

- Random Forest Classifier: An ensemble method that operates by constructing multiple decision trees during training. For classification, the output is the class chosen by the majority of the trees. It reduces overfitting by creating randomness in feature and data selection for each tree.

- Gradient Boosting Classifier: An ensemble technique that builds a series of weak learners (typically decision trees) sequentially. Each new tree corrects the errors of its predecessor, and the final prediction is a weighted sum of the predictions from all the trees.

- eXtreme Gradient Boosting (XGBoost): An optimized and highly efficient implementation of gradient boosting. It is known for its speed and performance and includes regularization techniques to prevent overfitting.

- LightGBM: An open-source gradient-boosting framework that uses a tree-based learning algorithm. It is designed to be highly efficient and scalable, particularly for large datasets, by using a technique called Gradient-based One-Side Sampling (GOSS).

## Ensemble Learning Model

Ensemble learning is a powerful machine learning paradigm where multiple models, known as "base learners," are trained to solve the same problem. The predictions from these individual models are then combined to produce a more accurate and robust final prediction than any single model could achieve on its own. The core principle is that a diverse group of "weak learners" can collectively form a "strong learner" by mitigating individual errors and biases. There are several key ensemble methods, with bagging (e.g., Random Forest) and boosting (e.g., Gradient Boosting, XGBoost, LightGBM) being the most prominent.

- Bagging (Bootstrap Aggregating): This method involves training multiple base models independently on different subsets of the data, which are created by sampling with replacement (bootstrapping). The final prediction is determined by averaging the predictions (for regression) or taking a majority vote (for classification). The randomness in the data sampling helps to reduce variance and prevent overfitting.

- Boosting: In contrast to bagging, boosting builds models sequentially. Each new model is trained to correct the errors made by the previous models. This process gives more weight to misclassified instances, forcing subsequent models to focus on the most difficult parts of the data. This approach is highly effective at reducing bias.

## Hyperparameter Tuning

**Theoretical Background**

Hyperparameter tuning is a crucial step in the machine learning workflow that aims to optimize a model's performance. Unlike model parameters, which are learned from the data (e.g., coefficients in a linear model), hyperparameters are external configuration settings that are set prior to the training process (e.g., the number of neighbors in k-NN or the learning rate in a boosting model). The goal of tuning is to find the combination of hyperparameters that yields the best possible performance on unseen data, thereby achieving the model's objective, such as maximizing accuracy or minimizing error.

One of the most common and systematic methods for hyperparameter tuning is a grid search. In a grid search, a predefined grid of hyperparameter values is created, and the model is trained and evaluated for every possible combination of values in the grid. This exhaustive search guarantees that the best combination of values within the specified range is found. While computationally expensive, it is a reliable method for exploring the parameter space.

### Support Vector Machine

**Theory Behind Support Vector Machine Hyperparameters**

Support Vector Machines (SVMs) are powerful models used for both classification and regression. The performance of an SVM is highly dependent on its hyperparameters, which allow for fine-tuning the model's behavior and complexity. These parameters control the trade-off between fitting the training data accurately and maintaining a simple, generalizable model.

**Important Parameters**

**C (Regularization Parameter):**

The C parameter is a positive value that controls the penalty for misclassified training data points. It represents a trade-off between achieving a smooth decision boundary and correctly classifying all training points.
- Small C: A small C value places less emphasis on correctly classifying individual data points. This results in a wider margin and a simpler, smoother decision boundary, which can reduce overfitting and improve generalization on unseen data.
- Large C: A large C value penalizes misclassifications more heavily, forcing the model to fit the training data as accurately as possible. This can lead to a narrower margin and a more complex decision boundary, which may result in overfitting.

**Kernel:**

The kernel is a function that transforms the input data into a higher-dimensional feature space. This allows the model to find a linear decision boundary even when the original data is not linearly separable. The choice of kernel is crucial as it determines the nature of the decision boundary.
- linear: Uses a linear function to create a straight decision boundary. This is suitable for data that is linearly separable.
- rbf (Radial Basis Function): Also known as the Gaussian kernel, it is effective for non-linear data. It maps data to an infinite-dimensional space, creating a complex, non-linear decision boundary.
- poly (Polynomial): Creates a non-linear decision boundary by using a polynomial function of a specified degree.
- sigmoid: Uses a sigmoid function and is a popular choice in neural networks.

**gamma:**

The gamma parameter, a kernel coefficient, is used with the rbf, poly, and sigmoid kernels. It defines the influence of a single training example on the model's decision boundary.
- Low gamma: A low gamma value means a large influence radius, causing points farther away from the decision boundary to be considered. This results in a smoother, less complex model.
- High gamma: A high gamma value means a small influence radius, causing only points very close to the decision boundary to be considered. This can lead to a highly complex, "spiky" decision boundary that may overfit the training data.

**degree:**

This parameter is exclusively used for the poly kernel and represents the degree of the polynomial function. A higher degree allows the model to fit more complex, curved relationships in the data.

**coef0:**

This is an independent term in the kernel function, which is relevant only for the poly and sigmoid kernels. It is used to adjust the model's sensitivity to high-dimensional data.

**tol (Tolerance):**

The tol parameter specifies the tolerance for the stopping criterion. The model will stop training when the change in the objective function is less than this value.

### eXtreme Gradient Boosting (XGBoost)

**Theory Behind eXtreme Gradient Boosting (XGBoost)**

eXtreme Gradient Boosting (XGBoost) is a highly efficient and scalable implementation of the gradient boosting framework. At its core, XGBoost is an ensemble learning method that builds a series of weak learners, typically decision trees, in a sequential manner. Each new tree is constructed to correct the errors of the previously built trees. This process is optimized through several key enhancements:

- **Regularization:** XGBoost includes both L1 (Lasso) and L2 (Ridge) regularization terms in its objective function. This helps to prevent overfitting by penalizing complex models and encouraging simpler, more generalizable trees.
- **Parallelism:** Unlike traditional boosting algorithms, XGBoost can parallelize the tree-building process, making it significantly faster and more scalable, especially on large datasets.
- **Handling Missing Values:** XGBoost has a built-in mechanism to handle missing values by automatically learning the best direction to send a data point with a missing value.
- **Smarter Splits:** It employs a more precise method for finding the optimal split points in a tree, leading to better-performing models.

**Important Hyperparameters for Grid Search**

The following hyperparameters are crucial for tuning an XGBoost model and were selected for the grid search:
- **n_estimators:** This parameter determines the number of boosting rounds or trees to be built. A higher value can improve performance but increases the risk of overfitting.
- **max_depth:** This controls the maximum depth of each tree. A deeper tree can model more complex relationships but may lead to overfitting, while a shallower tree can cause underfitting.
- **eta (learning rate):** The learning rate determines the step size at each boosting iteration. A smaller eta requires more boosting rounds but can lead to a more robust model.
- **subsample:** This specifies the fraction of the training instances that will be randomly sampled for each tree. It can help reduce variance and prevent overfitting.
- **colsample_bytree:** This controls the fraction of features that will be randomly sampled for each tree. A lower value can help prevent overfitting by making the trees more diverse.
- **gamma:** This parameter sets the minimum loss reduction required to make a further partition on a leaf node. A larger gamma results in a more conservative algorithm.
- **reg_alpha (L1 regularization):** This adds a penalty to the objective function proportional to the absolute value of the coefficients. A larger value makes the model more conservative, helping to prevent overfitting.
- **reg_lambda (L2 regularization):** This adds a penalty to the objective function proportional to the square of the coefficients. Similar to reg_alpha, a larger value makes the model more conservative.

**Model Assessment**

The final assessment of the model's performance is a critical step in the machine learning workflow. Given that this is a classification model, the evaluation metrics must be chosen accordingly. Metrics commonly used for regression tasks, such as Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared (R2), are not suitable for this problem.

Instead, the model's performance will be evaluated using a comprehensive suite of classification metrics:

- **Confusion Matrix:** This visual summary of the prediction results shows the number of correct and incorrect predictions for each class, providing insights into the types of errors the model is making.
- **Accuracy, Precision, Recall, and F1-Score:** These metrics provide a holistic view of the model's predictive capability.
  - Accuracy measures the overall proportion of correct predictions.
  - Precision measures the proportion of positive identifications that were actually correct.
  - Recall measures the proportion of actual positives that were identified correctly.
  - The F1-Score is the harmonic mean of precision and recall, providing a balanced measure of the model's performance, especially in cases of class imbalance.

**Cost-Sensitive Accuracy** 

This metric is particularly useful when the costs of different types of errors are not equal. It allows for an evaluation that weights certain mistakes more heavily, aligning the model's performance with real-world business objectives.

## ROC Curve and Area under ROC Curve (AUC)

The ROC curve plots the model's True Positive Rate (TPR) against its False Positive Rate (FPR). A model that performs well will have a curve that bows towards the top-left corner of the plot, indicating a high TPR while maintaining a low FPR. 

- TPR close to 1 while keeping FPR near 0.
- more area under the ROC Curve (AUC) indicated that obtain a good classifier.

```python
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
```

```python
# Assuming predicted probabilities for multiple models
yt_XGB, yp_XGB, _ = roc_curve(y_test, y_predicted_XGB)
roc_auc_XGB = auc(yt_XGB, yp_XGB)

# Plot ROC curves for multiple models
plt.figure(figsize=(8, 6))
plt.plot(yt_XGB, yp_XGB, color='lightcoral', lw=2, label='XGBoost Classifier' % roc_auc_XGB)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison of Models')
plt.legend(loc="lower right")
plt.show()
```
